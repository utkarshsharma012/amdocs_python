```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,color='g',linewidth=4,linestyle='dotted')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('2-D Plots')
plt.grid()
plt.show()
```

# Multiple Plots In one figure

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,44,34,37,23,33,41]
avg_t=[45,48,48,46,40,41,41]
```

```python
plt.plot(days,max_t,label="Max Temp")
plt.plot(days,min_t,label="Min Temp")
plt.plot(days,avg_t,label="Avg Temp")

plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('2-D Plots')
plt.legend(loc='best')
plt.grid()
plt.show()
```

# Sub Plots

```python

# plt.subplot(row,columns,loc_plot)
plt.subplot(2,2,1)
plt.plot(days,max_t,label="Max Temp")
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('2-D Plots')
plt.legend(loc='best')
plt.grid()
plt.subplot(2,2,2)
plt.plot(days,min_t,label="Min Temp")
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('2-D Plots')
plt.legend(loc='best')
plt.grid()
plt.subplot(2,2,4)
plt.plot(days,avg_t,label="Avg Temp")
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('2-D Plots')
plt.legend(loc='best')
plt.grid()


plt.show()
```

# Bar Chart's 

```python
company=['Google','AMZN','MSFT','XioMI']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue,label="Revenue(Bln)")

plt.ylabel('Revenue(Bln)')
plt.title('Us Tech Stocks')
plt.legend(loc='best')
plt.grid()


plt.show()
```

```python
plt.barh(company,revenue,label="Revenue(Bln)")

plt.xlabel('Revenue(Bln)')
plt.title('Us Tech Stocks')
plt.legend(loc='best')
plt.grid()


plt.show()
```

```python
profit=[40,41,34,12]
```

# Stacked Bar Chart

```python
plt.bar(company,revenue,label="Revenue(Bln)")
plt.bar(company,profit,label="Profit")

plt.ylabel('Revenue(Bln)/Profit')
plt.title('Us Tech Stocks')
plt.legend(loc='best')
plt.grid()


plt.show()
```

# Grouped Bar Chart

```python
import numpy as np
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2,revenue,label="Revenue(Bln)",width=0.4)
plt.bar(xpos+0.2,profit,label="Profit",width=0.4)

plt.ylabel('Revenue(Bln)/Profit')
plt.xticks(xpos,company)
plt.title('Us Tech Stocks')
plt.legend(loc='best')
plt.grid()


plt.show()
```

# Pie Chart

**Draw pie chart to track down home expenses**

```python
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','Car','Others']
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.2f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.2f%%',explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.2f%%',explode=[1,1,1,1,1])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.2f%%',explode=[0,0,0,0.2,1])
plt.show()
```
